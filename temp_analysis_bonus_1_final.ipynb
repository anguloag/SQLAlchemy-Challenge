### Compare June and December data across all years 

```python
from scipy import stats

# Reset the index of hawaii_df to call on Date column
hawaii_df = hawaii_df.reset_index()
hawaii_df
```

```python
# Filter data for desired months (June and December)
hawaii_june = hawaii_df.loc[(hawaii_df['date'].dt.month==6), :]
hawaii_june
```

```python
hawaii_dec = hawaii_df.loc[(hawaii_df['date'].dt.month==12), :]
hawaii_dec
```

```python
# Identify the average temp for June at all stations across all available years. Repeat for December.
# ---have to have same number of values for each sample when conducting the ttest
# ---since June has more rows than December, group the data by station and find the mean temp for each

june_avg_byStation = hawaii_june['tobs'].groupby(hawaii_june['station']).mean()
print(june_avg_byStation)

dec_avg_byStation = hawaii_dec['tobs'].groupby(hawaii_dec['station']).mean()
print(dec_avg_byStation)
```

```python
# Create collections of temperature data -- set the avg.s by station to lists
june_avg_byStation_list = june_avg_byStation.to_list()
print(june_avg_byStation_list)

dec_avg_byStation_list = dec_avg_byStation.to_list()
print(dec_avg_byStation_list)
```

```python
# Run paired t-test (Time 1=June, Time 2=Dec)
stats.ttest_rel(june_avg_byStation_list, dec_avg_byStation_list)
```

### Analysis

We run a paired t-test because we are comparing the temperatures at two different times periods (June and December) but for the same location (Hawaii); therefore, the June and December samples are related. 
The results of the paired t-test show that the difference in mean temperature in Hawaii from June and December are statistically significant, with a p-value of 0.0001.
